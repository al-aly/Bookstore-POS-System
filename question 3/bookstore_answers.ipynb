```python
# extablish connection with server
def create_server_connection(host_name, user_name, user_password):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection

pw = "Kud@+3rb@ng" 
db = "bookstore" 

connection = create_server_connection("localhost", "root", pw)
```

```python
# extablish connection with database
def create_db_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
```

```python
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query successful")
    except Error as err:
        print(f"Error: '{err}'")
```

```python
def read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as err:
        print(f"Error: '{err}'")
```

Question 2(b) : Show the SQL query for number of customers purchasing more than 5 books

```python
# Returns a list of lists and then creates a pandas DataFrame
from_db = []

question_b = """
SELECT 
customers.name AS customers_name,
SUM(invoice_lines.quantity)  AS quantity,
SUM(invoice_lines.total)
FROM customers 

JOIN invoices
ON invoices.customer_id = customers.id

JOIN invoice_lines
ON invoice_lines.invoice_id = invoices.id

GROUP BY customers.name
HAVING quantity > 5

"""

connection = create_db_connection("localhost", "root", pw, db)
results = read_query(connection, question_b)

for result in results:
    result = list(result)
    from_db.append(result)
 

    columns = ["name", "quantity", "total"]
df = pd.DataFrame(from_db,columns=columns)


display(df)
```

```python
# Returns a list of lists and then creates a pandas DataFrame
from_db = []

question_c = """
SELECT 
customers.id,
customers.name AS customers_name
FROM customers 

WHERE  customers.id NOT IN (SELECT invoices.customer_id FROM invoices)
"""

connection = create_db_connection("localhost", "root", pw, db)
results = read_query(connection, question_c)

for result in results:
    result = list(result)
    from_db.append(result)
 

    columns = ["id","name"]
df = pd.DataFrame(from_db,columns =columns)


display(df)
```

 Question 2(d) : Show the SQL query for list of book purchased with the users

```python
# Returns a list of lists and then creates a pandas DataFrame
from_db = []

question_d = """
SELECT 
invoice_lines.description AS books,
invoice_lines.quantity AS quantity,
customers.name AS customers_name
FROM customers 

INNER JOIN invoices
ON invoices.customer_id = customers.id

LEFT JOIN invoice_lines
ON invoice_lines.invoice_id = invoices.id

ORDER BY books
"""

connection = create_db_connection("localhost", "root", pw, db)
results = read_query(connection, question_d)

for result in results:
    result = list(result)
    from_db.append(result)
 

    columns = ["description","quantity", "name"]
df = pd.DataFrame(from_db,columns=columns)


display(df)
```
